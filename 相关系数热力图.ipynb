```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('zhengqi_train.txt',sep='\t')
test_data = pd.read_csv('zhengqi_test.txt',sep='\t')
```

**<font size = 4 color="#414141">相关系数热力图</font>**

```python
corf = np.corrcoef(train_data,rowvar=0)#返回nparray类型
```

```python
colnm = train_data.columns.tolist()
mcorr = train_data[colnm].corr(method="spearman")#返回dataframe类型
```

```python
corf
```

```python
mcorr
```

```python
plt.figure(figsize = (20,16))
mask = np.zeros_like(mcorr,dtype=np.bool)
mask[np.triu_indices_from(mask)] = True#去掉另一边triu_indices_from返回上三角形的索引。
cmap = sns.diverging_palette(220,10,as_cmap=True)#as_cmap简单来讲是否渐进
g = sns.heatmap(mcorr,mask = mask,cmap = cmap,square = True, annot = True, fmt = '0.2f')#square设置热力图矩阵小块形状，annot是否在写入数据，fmt数据格式比如保留几位小数
plt.show()
```
